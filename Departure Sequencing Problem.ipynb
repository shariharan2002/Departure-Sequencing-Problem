# Departure Sequencing Problem

#### 1. Get the k value

```python
N=3
kmax=[5,5,5]
k=[1,2,3]

```

#### 2. Define time separation matrix

```python
tij_list=[[70,100,130],[70,80,110],[70,80,90]]
tij={}
for i in range(len(tij_list)):
    for j in range(len(tij_list)):
        tij[(tuple((k[i],k[j])))]=tij_list[i][j]
```

#### 3. Define the Objective Z Value

```python
Z=1
#PCount=[150,120,100]
#Z=2
```

#### 4. Define the V and NEXT Dictionary in lexicographic increasing manner, with value =0

```python
def generate_combinations(kmax):
    num_variables = len(kmax)
    combinations = []

    def generate_recursive_combinations(variable_index, current_combination):
        if variable_index == num_variables:
            combinations.append(current_combination.copy())
        else:
            for k in range(kmax[variable_index] + 1):
                current_combination[variable_index] = k
                generate_recursive_combinations(variable_index + 1, current_combination)

    initial_combination = [0] * num_variables
    generate_recursive_combinations(0, initial_combination)

    return combinations
k_combinations=generate_combinations(kmax)
V={}
NEXT={}
for combination in k_combinations:
    for category in k:
        temp=[category]+combination
        V[tuple(temp)]=0
        NEXT[tuple(temp)]=0
targets_to_calculate=list(V.keys())
for i in range(len(kmax)):
    targets_to_calculate.pop(0)
```

##### 5. Perform the Actual Dynamic Programming Step

```python
for i in range(len(targets_to_calculate)):
    minner=10**8
    minner_guy=(-1,-1,-1,-1)
    for category in k:
        #print(category)
        relocate_subproblem=list(targets_to_calculate[i])
        if relocate_subproblem[category]>=1:
            new_guy=relocate_subproblem.copy()
            new_guy[category]-=1
            new_guy[0]=category
            new_guy=tuple(new_guy)
            if Z*(tij[tuple((relocate_subproblem[0],category))])+V[new_guy]<minner:
                minner_guy=new_guy
                minner=Z*(tij[tuple((relocate_subproblem[0],category))])+V[new_guy]
    V[targets_to_calculate[i]]=minner
    NEXT[targets_to_calculate[i]]=minner_guy
    
        
    
```
